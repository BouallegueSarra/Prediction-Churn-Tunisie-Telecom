# Prétraitement des données

* **Suppression des colonnes inutiles**


```python
for x in data.columns:
    print(x , len(data[x].unique()))
```

la colonne "id_client" contient 5000 valeurs uniques, donc il serait préférable de la supprimer de notre base de données.

```python
data = data.drop(['id_client'], axis= 1)
```

```python
data.head()
```

* **Les valeurs manquantes**

```python
# vérifier les valeur null
data.isnull().sum()
```

Il y a des valeurs manquantes au niveau de 9 colonnes.

```python
plt.figure(figsize=(15, 7))
sns.heatmap(data.isna(), cbar=False)
plt.show()
```

**Traitement de la colonne "cout_appel_jour"**

On a remarqué que: 
- la colonne "cout_appel_jour" est le coût total des minutes d'appel effectué par jour indiqué dans la colonne "durée_appel_jour(minutes)" . 
D'ou on a calculé le coût d'une minute par jour qui est 265,1/45.07 = 5.88

- Maintenant par ce coût d'une minute, on va remplir la colonne "cout_appel_jour" qui est 5.88*durée d'appel par jour (minutes)

```python
data["cout_appel_jour"] = data["cout_appel_jour"].fillna(data["durée_appel_jour(minutes)"] * 5.88)
```

**Traitement de la colonne "nb_reclamation"**

```python
data["nb_reclamation"].value_counts()
```

```python
data["nb_reclamation"] = data["nb_reclamation"].fillna(1.0)
```

Remplacer les valeur manquantes de colonne "nb_reclamation" par la valeur la plus élevée observée dans cette colonne, qui est de 1.

**Traitement de la colonne "nb_appel_inter"**

```python
data["nb_appel_inter"].value_counts()
```

```python
mean_value = data["nb_appel_inter"].mean()
mean_value= round(mean_value)
mean_value
```

```python
data["nb_appel_inter"] = data["nb_appel_inter"].fillna(mean_value)
```

**Traitement de la colonne "nb_appel_inter", "nb_appel_jour", "nb_appel_soirée", "nb_appel_nuit"**

```python
data["nb_jours_abonne"].value_counts()
```

```python
data["nb_appel_soirée"].value_counts()
```

```python
data["nb_appel_nuit"].value_counts()
```

```python
data["nb_appel_jour"].value_counts()
```

```python
data["nb_jours_abonne"] = data["nb_jours_abonne"].fillna(data["nb_jours_abonne"].max())
data["nb_appel_jour"] = data["nb_appel_jour"].fillna(data["nb_appel_jour"].max())
data["nb_appel_soirée"] = data["nb_appel_soirée"].fillna(data["nb_appel_soirée"].max())
data["nb_appel_nuit"] = data["nb_appel_nuit"].fillna(data["nb_appel_nuit"].max())
```

**Traitement de la colonne "active_msg_vocaux"**

```python
data["active_msg_vocaux"].value_counts()
```

```python
data["active_msg_vocaux"].fillna(method="ffill", inplace=True)
```

Utiliser la méthode ffill (forward fill) pour remplacer la valeur manquante dans la colonne "active_msg_vocaux" par la dernière valeur observée avant la valeur manquante.

**Traitement de la colonne "churn"**

```python
data["churn"].value_counts()
```

```python
data["churn"].fillna(method="bfill", inplace=True)
```

Utiliser la méthode bfill (backward fill) pour remplacer la valeur manquante dans la colonne "churn" par la première valeur observée après la valeur manquante.

**Vérification des valeurs manquantes**

```python
data.isnull().sum()
```

Après avoir effectué le traitement, il ne reste plus de valeurs manquantes dans le dataset.

# Normalisation

```python
x=data.drop(["churn"], axis=1)
y=data["churn"]
x
#axis=1 : l'axe des colonnes.
```

* variable prédictive : Toute les variables sauf "churn"
* variable cible  : Churn

```python
#np.random.seed(0)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0, stratify=y)
print("X_train : ",x_train.shape) 
print("X_test : ",x_test.shape) 
print("Y_train : ",y_train.shape) 
print("Y_test : ",y_test.shape)
```

On divise les données en 2 parties :
   * 80% : L'entrainement des données
   * 20% : Le test de modèle

```python
sc=StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

StandardScaler() modifie les valeurs des colonnes numériques pour utiliser une échelle commune centrée autour de 0. 

z = (x - u) / s
* x : la valeur de la variable à transformer
* u : la moyenne de la variable
* s : l'écart type de la variable
* z : la valeur transformée de la variable

# Modeling

### 1- Logistic Regression


```python
log_regression = LogisticRegression(random_state=0)
log_regression.fit(x_train,y_train)
score_logisticRegression = log_regression.score(x_test, y_test)*100
print(" Le test score est = ", score_logisticRegression)
```

* **Matrice de confusion**

```python
y_pred = log_regression.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
y_labels = ['non churn (0)', 'churn (1)']
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, cmap="Blues" )
plt.xlabel('Classe prédite')
plt.ylabel('Classe réelle')

plt.gca().set_xticklabels(y_labels)
plt.gca().set_yticklabels(y_labels)

plt.show()
```

* **Classification report**


```python
print(classification_report(y_test, y_pred, digits=5))
```

*  **Learning Curve**


```python
train_size, train_score, test_score = learning_curve(log_regression, x_train, y_train, train_sizes=np.linspace(0.1,1,10), cv=5)
plt.plot(train_size, train_score.mean(axis=1), train_size, test_score.mean(axis=1))
plt.legend(('train_score', 'test_score'))
plt.xlabel('train_size')
plt.ylabel('score')
plt.grid(True)
plt.show()
```

```python
params = {
        'min_child_weight': range(10),
        'gamma': [0.5, 1, 1.2],
        'subsample': [0.6, 0.8],
        'colsample_bytree': [0.5, 0.6, 0.7, 0.8, 0.9, 1.0, 1.1, 1.2, 1.3, 1.4, 1.5],
        'max_depth': range(10)
        }
grid = RandomizedSearchCV(XGBClassifier(), param_distributions=params, cv =5, n_iter=10)
grid.fit(x_train, y_train)
print("Les meilleurs hyperparamètres :",grid.best_params_) 
```
