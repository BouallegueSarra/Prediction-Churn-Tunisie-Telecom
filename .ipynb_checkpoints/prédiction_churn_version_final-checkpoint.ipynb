## Modeling

```python
x=data.drop(["churn"], axis=1)
y=data["churn"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=5)
print("X_train : ",x_train.shape) 
print("X_test : ",x_test.shape) 
print("Y_train : ",y_train.shape) 
print("Y_test : ",y_test.shape)
```

###  Logistic Regression

```python
#normalisation 
sc=StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
log_regression = LogisticRegression()
log_regression.fit(x_train,y_train)

test_score = log_regression.score(x_test, y_test)
train_score = log_regression.score(x_train, y_train)
print(" train score : ", train_score)
print(" test score : ", test_score)
```

```python
# matrice de confusion
plot_confusion_matrix(log_regression, x_test, y_test,cmap=plt.cm.Blues)
plt.show()
y_prediction = log_regression.predict(x_test)
print(classification_report(y_test, y_prediction, digits=6))
```

```python
train_size, train_score, test_score = learning_curve(log_regression, x_train, y_train, train_sizes=np.linspace(0.1,1,10), cv=5, random_state=0)
plt.plot(train_size, train_score.mean(axis=1), train_size, test_score.mean(axis=1))
plt.legend(('train_score', 'test_score'))
plt.xlabel('train_size')
plt.ylabel('score')
plt.grid(True)
plt.show()
```
