### 5.3 Feature selection 

```python
def strongcorr (m,y):
  l=m.shape[1]
  fs= set()
  for i in range(l):
    max=i
    min=i
    for j in range(i):
      if(abs(m.iloc[i,j]>0.6)):
        if(abs(y[j][0])>abs(y[i][0])):
          min=max
          max=j
        else:
          min=j
        fs.add(m.columns[min])
  return(fs)
```

```python
df1=df.drop(['Churn Value'], axis=1)
m=df1.corr()
y=df.corr()[['Churn Value']]
y=y[ y['Churn Value']!=1].to_numpy()
strongcorr(m,y)
```

```python
def weakcorr (Y):
  f2= set()
  for i in range(Y.shape[0]):
    if abs(Y.iloc[i][0])<0.1 :
      f2.add(Y.index[i])
  return(f2)
```

### xgBoost

```python
#Splitting our data into training and testing set
y=df['Churn Value']
x=df.drop(['Churn Value'], axis=1)
x1_train,x1_test,y1_train,y1_test=train_test_split(x,y,test_size=0.3,random_state=20)
```

```python
# min max scaler application
sc=MinMaxScaler()
x1_train=sc.fit_transform(x1_train)
x1_test=sc.transform(x1_test)
```

```python
#Model training and evaluating
modelxg = XGBClassifier()
modelxg.fit(x1_train, y1_train)
print('Le train score est :', modelxg.score(x1_train, y1_train))
print('Le test score est :', modelxg.score(x1_test, y1_test))
```

```python
# we will be using randomizedSearchCV because its way faster than  GridSearchCV
params = {
        'min_child_weight': range(10),
        'gamma': [0.5, 1, 1.2],
        'subsample': [0.6, 0.8],
        'colsample_bytree': [0.5, 0.6, 0.7, 0.8, 0.9, 1.0, 1.1, 1.2, 1.3, 1.4, 1.5],
        'max_depth': range(10)
        }
grid = RandomizedSearchCV(XGBClassifier(), param_distributions=params, cv =5, n_jobs=-1, n_iter=10)
grid.fit(x1_train, y1_train)
grid.best_params_ #printing best parametres
```

```python
#Using the hyperparameters suggested by RandomizedSearchCV
modelxg_final = XGBClassifier(subsample=0.8 , min_child_weight=8,max_depth=3,gamma=1.2, colsample_bytree=1.0 ,random_state=5)
modelxg_final.fit(x1_train, y1_train)
print('Le train score est :', modelxg_final.score(x1_train, y1_train))
print('Le test score est :', modelxg_final.score(x1_test, y1_test))
```

```python
#confusion matrix 
plot_confusion_matrix(modelxg_final, x1_test, y1_test)
plt.show()
```

```python
#Checking if our model will perform better by increasing the amount of training data
train_size, train_score, test_score = learning_curve(modelxg_final, x1_train, y1_train, train_sizes=np.linspace(0.1,1,50), cv=10, random_state=3 )
plt.plot(train_size, train_score.mean(axis=1), train_size, test_score.mean(axis=1))
plt.legend(('train_score', 'test_score'))
plt.xlabel('train_size')
plt.ylabel('score')
plt.grid(True)
plt.show()
```

Increasing the amount of data will not affect the overall accuracy
