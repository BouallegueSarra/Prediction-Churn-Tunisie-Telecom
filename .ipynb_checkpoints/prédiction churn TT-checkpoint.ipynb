## Préparation des données

## Encodage des données


```python
label_encoder = LabelEncoder()
data['active_msg_vocaux'] = label_encoder.fit_transform(data['active_msg_vocaux'])
data['churn'] = label_encoder.fit_transform(data['churn'])
```
